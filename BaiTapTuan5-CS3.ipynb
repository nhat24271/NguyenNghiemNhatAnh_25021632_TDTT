```python
#W5A4
def ktra(n):
    if n<=1:
        return False
    cnt=0
    for i in range(1, (n//2)+1):
        if n%i==0:
            cnt+=i
    return cnt==n
a=int(input('nhap so: '))
if ktra(a):
    print('so hoan hao')
else:
    print('k la so hoan hao')
```

```python
#W5A5
def pos_k_in_list(mylist: list, k: int) -> int:
    for i in range(len(mylist)):
        if mylist[i]==k:
            return i+1
    return -1
mylist=list(map(int, input().split()))
k=int(input())
print(pos_k_in_list(mylist,k))
```

```python
#W5A6
def giai_thua(n):
    if n==0 or n==1:
        return 1
    else:
        return n*giai_thua(n-1)
while True:
    try:
        n=int(input())
        if n>=0:
            print(giai_thua(n))
            break
        else:
            print('vui long nhap so khong am')
    except ValueError:
        print('chi nhap so nguyen')
```

```python
#W5A7
def may_tinh(num1, num2, operat):
    kq=0
    if operat=='+':
        kq=num1+num2
    elif operat=='-':
        kq=num1-num2
    elif operat=='*':
        kq=num1*num2
    elif operat=='/':
        if num2==0:
            print('loi phep toan')
            return
        else:
            kq=num1/num2
    print(f'ket qua cua phep {operat} 2 so tren la: {kq:.2f}')
print('nhap 2 so a,b va toan tu:')
try:
    a,b = map(float, input('nhap 2 so a,b: ').split())
    op = input('nhap toan tu: ')
    may_tinh(a, b, op)
except ValueError:
    print('vui long nhap dung dinh dang')

    
```

```python
#W5A8
def hamming(a, b):
    kq=a^b
    cnt=0
    while kq>0:
        if kq&1:
            cnt+=1
        kq=kq>>1
    return cnt
a,b=map(int, input().split())
print(hamming(a,b))
```

```python
#W5A10
def dang_cau(a, b):
    ax_a_den_b={}
    ax_b_den_a={}
    for i in range(len(a)):
        char_a=a[i]
        char_b=b[i]
        if char_a in ax_a_den_b:
            if ax_a_den_b[char_a]!=char_b:
                return False
        else:
            if char_b in ax_b_den_a:
                return False
        ax_a_den_b[char_a]=char_b
        ax_b_den_a[char_b]=char_a
    return True
a, b = input().split()
print('True' if dang_cau(a, b) else 'False') 
```
