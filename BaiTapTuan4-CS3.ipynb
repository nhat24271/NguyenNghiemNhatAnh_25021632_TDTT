```python
#W4A8
try:
    n=int(input())
    if n<0:
        print('vui long nhap so khong am')
    else:
        a=n
        step=0
        while True:
            temp_n=n
            reverse=0
            while temp_n>0:
                so_cuoi=temp_n%10
                reverse=reverse*10 + so_cuoi
                temp_n//=10
            if n==reverse:
                break
            n=n+reverse
            step+=1
        print(f'so doi xung cua {a} la: {n} va trai qua {step} buoc chay.')
except ValueError:
    print('vui long nhap so nguyen')
        
                
```

```python
#W4A8
try:
    n=int(input())
    while n<=0:
        print('vui long nhap so nguyen duong')
        n=int(input())
    ds=[]
    i=1
    while i*i<n:
        sq=i*i
        ds.append(sq)
        i+=1
        kq=[]
    for so in ds:
        chuoi=str(so)
        sap_xep=set(chuoi)
        if len(chuoi)==len(sap_xep):
            kq.append(so)
    if len(kq)==0:
        print('no number')
    else:
        for so in kq:
            print(so, end=' ')
except ValueError:
    print('chi nhap so nguyen.')
```

```python
#W4A13
a, b =map(int, input().split())
sum_uoc_a=0
sum_uoc_b=0
i=1
while i<=a//2:
    if a%i==0:
        sum_uoc_a+=i
    i+=1
j=1
if sum_uoc_a==b:
    while j<=b//2:
        if b%j==0:
            sum_uoc_b+=j
        j+=1
    print('True' if sum_uoc_b==a else 'False')
else:
    print('False')
```

```python
#W4A14
a, b = map(int, input().split())
max_uoc=0
i=1
while i<=min(a,b):
    if a%i==0:
        if b%i==0:
            max_uoc=i
    i+=1
print(f'uoc chung lon nhat cua {a} va {b} la: {max_uoc}')
```

```python
#W4A15
con, chan = map(int, input().split())
if chan%2!=0:
    print('invalid')
else:
    x=2*con - chan//2
    y=con - x
    if x>=0 and y>=0:
        print(f'so con ga va cho la: {x} va {y}')
    else:
        print('invalid')
```

```python
#W4A15
con, chan=map(int, input().split())
ga=0
tim_thay=False
while ga<=con:
    cho=con-ga
    tong_chan=ga*2 + cho*4
    if tong_chan==chan:
        print(f'{ga} {cho}')
        tim_thay=True
        break
    ga+=1
if not tim_thay:
    print('invalid')
```
