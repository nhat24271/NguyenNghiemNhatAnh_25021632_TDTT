```python
#W3A11
try:
    a, b, c = map(float, input('nhap do dai 3 canh cua tam giac: ').split())
    if a+b>c and a+c>b and b+c>a:
        if a==b or a==c or b==c:
            if a==b and a==c:
                print('day la tam giac deu.')
            else:
                print('day la tam giac can.')
        ds=[a, b, c]
        ds.sort()
        a=ds[0]
        b=ds[1]
        c=ds[2]
        if a**2 + b**2 == c**2:
            print('day la tam giac vuong.')
        else:
            print('day la tam giac thuong')
    else:
        print('day k phai do dai cua 1 tam giac.')
    
except ValueError:
    print('chi nhap so.')
```
